## Импорт библиотек

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## 1.1 Парсинг данных

```python
df=pd.read_csv('https://raw.githubusercontent.com/owid/covid-19-data/master/public/data/owid-covid-data.csv')
```

```python
df
```

### Проверка на наличие пустых значений

```python
pd.set_option('display.max_rows',None)
df.isnull().sum()
```

### Заполнение всех пустых значений

```python
pd.set_option('display.max_rows',5)
df=df.fillna(0)
```

### Добавление дополнительных атрибутов

#### Среднее количество новых заражений и новых смертей, сгруппированных по региону

```python
mean_cases_deaths=df[['location','new_cases',
                          'new_deaths']].groupby(by="location").mean().rename(columns={
                                                                                    'new_cases':'mean_cases', 
                                                                                    'new_deaths':'mean_deaths'})
df=df.merge(mean_cases_deaths, on='location')
```

```python
df
```

## 1.4. Формирование дополнительных атрибутов

Rt = число зарегистрированных заболеваний за последние 4 дня / число зарегистрированных заболеваний за предыдущие 4 дня

```python
df['Rt']=None
data=pd.DataFrame()
for country in df['location'].value_counts().keys():
    r=df[df['location']==country].copy()
    da=pd.DataFrame()
    for i in range(0, len(r), 8):
        tida=pd.DataFrame()
        su=r['new_cases'].tail(8).tail(4).sum()/r['new_cases'].tail(8).head(4).sum()
        tida=r.tail(8)
        tida['Rt']=su
        r.drop(r.tail(8).index,inplace=True)
        da=da.append(tida)
    data=data.append(da)
```

```python
data=data.fillna(0)
```

```python
data.reset_index(drop=True, inplace=True)
df=data
```
